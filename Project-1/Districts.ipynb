```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('District_Dataset.csv')
```

```python
df.head()
```

```python
plt.figure(figsize=(18,18))
plt.bar(df.District,df.Confirmed,color='grey',label='Confirmed Cases')
plt.bar(df.District,df.Recovered,color="green",label='Cured Cases')
plt.bar(df.District,df.Death,color="red",label='Death Cases')
plt.bar(df.District,df.Active,color="blue",label='Active Cases')
plt.title("COVID-19 District Wise Analysis of Maharashtra")
plt.legend()
plt.xticks(df.District,rotation=90)
plt.xlabel("District",labelpad=10)
plt.ylabel("Number of Cases",labelpad=10)
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.plot(df.District,df.Confirmed,color="black",label='Confirmed Cases')
plt.plot(df.District,df.Recovered,color="green",label='Cured Cases')
plt.plot(df.District,df.Death,color="red",label='Death Cases')
plt.plot(df.District,df.Active,color="blue",label='Active Cases')
plt.legend()
plt.xticks(rotation=90)
plt.xlabel("District",labelpad=10)
plt.ylabel("Number of Cases",labelpad=10)
plt.show()
```
