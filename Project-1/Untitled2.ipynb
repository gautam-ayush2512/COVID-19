```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('states.csv')
```

```python
df.head()
```

```python
plt.bar(df.State,df.Confirmed,label='Confirmed Cases',color='grey')
plt.bar(df.State,df.Recovered,color="green",label='Cured Cases')
plt.bar(df.State,df.Death,color="red",label='Death Cases')
plt.bar(df.State,df.Active,color="blue",label='Active Cases')
plt.title("COVID-19 ANALYSIS MAHARASHTRA & NEIGHBOUR:JUNE 2020")
plt.legend()
plt.xticks(df.State,rotation=90)
plt.xlabel("State",labelpad=10)
plt.ylabel("Number of Cases",labelpad=10)
plt.show()
```

```python
df.plot(y=['Confirmed','Active','Recovered','Death'], x='State',rot=90,figsize=(4,3))
```

```python
plt.pie(df.Active, labels=df.State)
plt.title("ACTIVE CASES :JUNE 2020")
```

```python
plt.pie(df.Confirmed, labels=df.State)
plt.title("CONFIRMED CASES :JUNE 2020")
```

```python
plt.pie(df.Recovered, labels=df.State)
plt.title("RECOVERED CASES :JUNE 2020")
```

```python
plt.pie(df.Death, labels=df.State)
plt.title("DEATH CASES :JUNE 2020")
```

```python
df.plot.bar(y=['Confirmed','Active','Recovered','Death'], x='State',figsize=(6,2))
plt.title("STATEWISE COVID-19 ANALYSIS : JUNE 2020")
```

```python
df.plot.bar(y=['CMR','IFR'], x='State',figsize=(6,2))
plt.title("STATEWISE IFR vs CMR : JUNE 2020")
```
