```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('complete1.csv')
```

```python
df.head()
```

```python
plt.figure(figsize=(10, 5))
plt.bar(df.Date,df.Confirmed,label='Confirmed Cases',color='lightskyblue')
plt.bar(df.Date,df.Recovered,color="palegreen",label='Cured Cases')
plt.bar(df.Date,df.New_deaths,color="black",label='Death Cases')
plt.legend()
plt.xticks(df.Date,rotation=90)
plt.xlabel("Date",labelpad=10)
plt.ylabel("Number of Cases",labelpad=10)
plt.show()
```

```python
plt.figure(figsize=(15, 8))
plt.plot(df.Date,df.Confirmed,color="orange",label='Confirmed Cases')
plt.plot(df.Date,df.Recovered,color="navy",label='Cured Cases')
plt.plot(df.Date,df.Death,color="red",label='Death Cases')
plt.legend()
plt.xticks(rotation=45)
plt.xlabel("Date",labelpad=10)
plt.ylabel("Number of Cases",labelpad=10)
plt.show()
```

```python
df.plot(y=['Recovered','Death','Confirmed'], x='Date',rot=90)
```
